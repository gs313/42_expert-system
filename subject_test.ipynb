# 🧠 Expert System — Evaluation Notebook

This notebook demonstrates all required features of the Expert System project.

It is structured according to the official evaluation sheet:
- Mandatory features
- Bonus features
- Error handling
- Additional edge cases

```python
from ExpertSystem import ExpertSystem
from parser import parse_string
import pandas as pd
```

```python
def run_test_string(content, expected):
    es = ExpertSystem()

    try:
        parse_string(content, es)
    except Exception as e:
        return {"error": str(e)}

    results = []

    for query, expected_result in expected.items():
        try:
            result = es.solve(query)
            results.append({
                "Query": query,
                "Expected": expected_result,
                "Got": result,
                "Pass": result == expected_result
            })
        except Exception as e:
            results.append({
                "Query": query,
                "Expected": expected_result,
                "Got": f"ERROR: {e}",
                "Pass": False
            })

    return pd.DataFrame(results)
```

```python
def show_test_string(title, content, expected):
    print(f"\n=== {title} ===")

    df = run_test_string(content, expected)

    if isinstance(df, dict):
        print("❌ ERROR:", df["error"])
        return

    display(df)

    score = df["Pass"].sum()
    total = len(df)

    print(f"Result: {score}/{total} passed")
```

```python
show_test_string(
    "AND Condition",
    """
    D + E => B
    B => A

    =DE
    ?A
    """,
    {
        "A": "T"
    }
)
```
